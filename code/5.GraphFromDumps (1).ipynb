# Get pagelinks from COVID Wikipedia related pages
This file reads the sample of the dumosp generated by the GetCovidPagesFromDump-Current.py script

```python
import json,gzip,glob
from tqdm import tqdm
import mwparserfromhell
import csv
```

```python
#Here using just the last version (current) from each article. To use the historical version remove the "current"
dumps = glob.glob('covid_pages_from_dump_current*.jsonl.gzip')
```

```python
for dump in dumps:
    lang = dump[-13:-11]
    f = gzip.open(dump,'rb')
    print(lang)
    with open('graphFromDump_current_%s.csv' % lang , 'w') as csvfile:
        csvwriter = csv.writer(csvfile, delimiter='\t',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
        csvwriter.writerow(['link_from_QID','link_from_title','link_to_title']) #write header
        for l in f:
            try:           
                data = json.loads(l.decode())
                link_from_title = data['page_title']
                link_from_pageid =data['page_id']
                link_from_QID = data['QID']
                #link_time = data['time']
                wikitext = mwparserfromhell.parse(data['text'])
                
                for link in wikitext.filter_wikilinks():
                    if ":" not in link.title: #filtering out special links like Template:
                        csvwriter.writerow([link_from_QID,link_from_title,link.title])
            except:
                print('error read',lang)
        
```
